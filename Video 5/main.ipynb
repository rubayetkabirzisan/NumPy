Numpy Arrays with random numbers

rand()

```python
import numpy as np
arr=np.random.rand(4)
print(arr)  
```

```python
var1=np.random.rand(2,5)
print(var1)
```

randn()

```python
var2=np.random.randn(4)
print(var2)
```

```python
var3=np.random.randn(4,5)
print(var3)
```

ranf()

```python
var4=np.random.ranf(4)
print(var4)
```

randint()

```python
var5=np.random.randint(1,100,5)
print(var5)
```

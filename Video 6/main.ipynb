Data types in numpy array

```python
import numpy as np
var=np.array([1,2,3,4,12,13,14,15])
print("Data type : ",var.dtype)
```

```python
var=np.array([1.0,1.2,1.3])
print("Data type : ",var.dtype)
```

```python
var=np.array(["a","s","t","r"])
print("Data type : ",var.dtype)
```

```python
var=np.array(["a","s","t","r",1,2,3,4])
print("Data type : ",var.dtype)
```

```python
x=np.array([1,2,3,4],dtype=np.int8)
print("Data type : ",x.dtype)
print(x)
```

```python
x1=np.array([1,2,3,4],dtype="f")
print("Data type : ",x1.dtype)
print(x1)
```

```python
x2=np.array([1,2,3,4])
new=np.float32(x2)
new1=np.int_(new)

print("Data type : ",x2.dtype)
print("Data type : ",new.dtype)
print("Data type : ",new1.dtype)

print(x2)
print(new)
print(new1)
```

```python
x3=np.array([1,2,3,4])
new_1=x3.astype(float)

print(x3)
print(new_1)
```

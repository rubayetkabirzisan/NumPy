The Concept of Matrix Numpy Arrays

```python
import numpy as np
var=np.matrix([[1,2,3],[1,2,3]])
var2=np.matrix([[1,2],[3,4],[1,2]])

print(var)
print(type(var))

print(var*var2)
print(var.dot(var2))
```

```python
var1=np.array([[1,2,3],[1,2,3]])
print(var1)
print(type(var1))
print(var1*var1)
```

Matrix Functions in NumPy

Transpose

```python
var3=np.matrix([[1,2,3],[4,5,6]])
print(var3)
print()
print(var3.transpose())
print(var3.T)
```

Swapaxes

```python
var4=np.matrix([[1,2,3],[4,5,6],[7,8,9]])
print(var4)
print()
print(var4.swapaxes(1,0))
```

Inverse

```python
var5=np.matrix([[1,2,3],[4,5,6],[7,8,8]])
print(var5)
print()
print(np.linalg.inv(var5))
print(var5.I)
```

Power

```python
var6=np.matrix([[1,2,3],[4,5,6],[7,8,8]])
print(var6)
print()
print(np.linalg.matrix_power(var6,2))
print()
print(np.linalg.matrix_power(var6,0))
print()
print(np.linalg.matrix_power(var6,-2))
print()
print(var6**2)
```

Determinant

```python
var7=np.matrix([[1,2,3],[4,5,6],[7,8,9]])
print(var7)
print()
print((np.linalg.det(var7)))
```

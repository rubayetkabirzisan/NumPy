Arithmetic Operations in Array

```python
import numpy as np

var = np.array([1,2,3,4])
varadd = var + 3
print(varadd)
```

Recirocal (Using Functions)

```python
var = np.array([1,2,3,4])
varadd = np.reciprocal(var)
print(varadd)
```

Adding Two One Dimensional Arrays(Other Arithmetic Operations Similarly)

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])

varadd = var1 + var2
print(varadd)
```

Using Via Functions

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])

varadd1 = np.mod(var1,var2)
print(varadd1)
```

Adding Two Dimensional Arrays(Other Arithmetic Operations Similarly)

```python
var21 = np.array([[1,2,3,4],[1,2,3,4]])
var22 = np.array([[1,2,3,4],[1,2,3,4]])

print(var21)
print()
print(var22)
print()

varadd2 = var21*var22
print(varadd2)

```

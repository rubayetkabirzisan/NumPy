Broadcasting in NumPy Arrays

```python
import numpy as np
var1=np.array([1,2,3])
print(var1)
print(var1.shape)
print()
var2=np.array([[1],[3],[4]])
print()
print(var2)
print(var2.shape)
print()
print(var2+var1)
```

```python
x=np.array([[1],[2]])
print(x.shape)

y=np.array([[1,2,3],[1,2,3]])
print(y.shape)
print((x+y))
```

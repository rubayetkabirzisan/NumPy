NumPy Arrays Functions-2

Shuffle

```python
import numpy as np
var=np.array([1,2,3,4,5])
np.random.shuffle(var)
print(var)
```

Unique

```python
var1=np.array([1,2,2,3,4,5,1])
x=np.unique(var1,return_index=True,return_counts=True)
print(x)
```

Resize

```python
var2=np.array([1,2,3,4,5,6])
y=np.resize(var2,(2,3))
print(y)
```

Flatten

```python
var3=np.array([1,2,3,4,5])
y=np.resize(var3,(3,2))
print(y)
print()
print(y.flatten(order="F"))#c,f,a,k
```

Ravel

```python
var3=np.array([1,2,3,4,5,6])
y=np.resize(var3,(3,2))
print(y)
print()
print(y.ravel(order="C"))#c,f,a,k
```

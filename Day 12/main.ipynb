
Iterating NumPy Arrays

One dimensional array

```python
import numpy as np
var = np.array([9,8,7,6,5,4])
print(var)
for i in var:
    print(i)
```

Two dimensional array

```python
var1=np.array([[1,2,3,4],[1,2,3,4]])
print(var1)
for i in var1:
    print(i)
print()
for k in var1:
    for j in k:
        print(j)
```

Three dimensional array

```python
var2=np.array([[[1,2,3,4]],[[1,2,3,4]]])
print(var2)
for i in var2:
    print(i)

print()
for k in var2:
    for j in k:
        for l in j:
            print(l)

print()

#nditer function

for i in np.nditer(var2,flags=['buffered'],op_dtypes=["S"]):
    print(i)
```

```python
#ndebunerate function
var3=np.array([[[1,2,3,4]],[[1,2,3,4]]])
print(var3)

print()

for i,d in np.ndenumerate(var3):
    print(i,d)
```

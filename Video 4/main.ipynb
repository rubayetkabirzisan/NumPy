Zeros Array

```python
import numpy as np
ar_zero=np.zeros(4)
ar_zero1=np.zeros((3,4))
print(ar_zero)
print()
print(ar_zero1)
```

Ones Array

```python
import numpy as np
ar_one=np.ones(4)
print(ar_one)
```

Empty Array

```python
ar_em=np.empty(4)
print(ar_em)
```

Range Array

```python
ar_rn=np.arange(4)
print(ar_rn)
```

Diagonal element filled with 1 Array

```python
ar_dia=np.eye(3)
print(ar_dia)
```

Array with values that are spaced linearly in a specified interval

```python
ar_lin=np.linspace(1,21,num=5)
print(ar_lin)
```

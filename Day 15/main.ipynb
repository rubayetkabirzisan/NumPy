NumPy Arrays Functions

Search

```python
import numpy as np

var=np.array([1,2,3,4,2,5,2,5,6,7])
var1=np.where(var%2==0)

print(var1)
```

```python
var_=np.array([1,2,3,4,6,7,8,9,10])
var1_=np.searchsorted(var_,[5,6,7],side="right")

print(var1_)
```

Sort

```python
var2=np.array([1,2,3,4,2,5,2,5,6,7])
var3=np.sort(var2)

print(var3)
```

```python
var2=np.array([[1,2,3],[4,2,5],[5,6,7]])
var3=np.sort(var2)

print(var3)
```

```python
var2_=np.array(["a","s","d","f"])
var3_=np.sort(var2_)

print(var3_)
```

Filter Array

```python
var4=np.array(["a","s","d","f"])
f=[True,False,False,True]
new_a=var4[f]

print(new_a)
```

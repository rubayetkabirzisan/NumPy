NumPy Insert and Delete Arrays Functions

```python
import numpy as np
var = np.array([1,2,3,4])
print(var)
# v=np.insert(var,(2,4),6.5)

x=np.append(var,6.5)
print()
print(x)
```

```python
var1=np.array([[1,2,3],[4,5,6]])
print(var1)
# v1=np.insert(var1,2,[11,12,13],axis=0)
v1=np.append(var1,[[11,12,13]],axis=0)
print()
print(v1)
```

Delete

```python
import numpy as np
var2 = np.array([1,2,3,4])
print(var2)
print()
d=np.delete(var2,2)
print(d)
```

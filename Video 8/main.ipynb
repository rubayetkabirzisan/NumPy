Arithmetic Functions In Numpy Arrays

```python
import numpy as np

var = np.array([1,2,3,4,6,2,3])
print("min : ",np.min(var))
print("max : ",np.max(var))
print("square root : ",np.sqrt(var))
print("min position : ",np.argmin(var))
print("max position : ",np.argmax(var))

```

Two Dimensional Arrays

```python
var1=np.array([[2,1,3],[9,5,6]])

print("min : ",np.min(var1))
print("max : ",np.max(var1))
print("min position : ",np.argmin(var1,axis=0))
# 0 for column, 1 for row
print("max position : ",np.argmax(var1,axis=0))
```

Trigonometric Functions

```python
var2=np.array([1,2,3])

print("sine : ",np.sin(var2))
print("cosine : ",np.cos(var2))
```

Cumulative Sum

```python
var3=np.array([1,2,3])

print("cumulative sum : ",np.cumsum(var3))
```

Joining & Split NumPy Arrays

```python
import numpy as np
var = np.array([1,2,3,4])
var1=np.array([5,6,7,8])
var2=np.concatenate((var,var1))

print(var)
print()
print(var1)
print()
print(var2)
```

```python
vr = np.array([[1,2],[3,4]])
vr1 = np.array([[5,6],[7,8]])
vr2=np.concatenate((vr,vr1),axis=0)

print(vr)
print()
print(vr1)
print()
print(vr2)
```

```python
vr_ = np.array([1,2,3,4])
vr1_ = np.array([5,6,7,8])

vr2_=np.stack((vr_,vr1_),axis=1)
vr3_=np.hstack((vr_,vr1_))
vr4_=np.vstack((vr_,vr1_))
vr5_=np.dstack((vr_,vr1_))

print(vr_)
print()
print(vr1_)
print()
print(vr2_)
print()
print(vr3_)
print()
print(vr4_)
print()
print(vr5_)
```

Split array

```python
var_=np.array([1,2,3,4,5,6])
print(var_)
print()
ar_=np.array_split(var_,2)
print(ar_[0])
print(type(ar_))
```

```python
var_1=np.array([[1,2],[3,4],[5,6]])
print(var_1)
print()
ar_1=np.array_split(var_1,3)
print(ar_1)
print()
ar_1=np.array_split(var_1,3,axis=1)
print(ar_1)
print(type(ar_1))
```
